```python
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
import re
```

```python
df = pd.read_csv('./genre_mapping.csv')
```

```python
METADATA_PATH = '../fma/fma_metadata/'
convertedGenres = pd.read_csv(METADATA_PATH + 'genres.csv')
genresTag = convertedGenres.loc[:, ['genre_id', 'title']]
```

```python
genreMap = dict(zip(genresTag['genre_id'],genresTag['title']))
pattern = '[\/\-\s\:]'
for i, j in genreMap.items():
    newName = j.lower()
    newName = re.sub(pattern, '_', newName)
    newName = re.sub('[\(\)]', '-', newName)
    newName = re.sub('&', 'and', newName)
    genreMap[i] = f'{i}-{newName}'
```

```python
gb = df.groupby('genre_id')
g = [gb.get_group(x) for x in gb.groups]
savePath = './track_by_genre_small'
musicSrcPath = '../fma/fma_small'
for i in g:
    csvName = 'none'
    data = {'category': [], 'path': []}
    for row, idx in i.iterrows():
        fileName = str(idx['track_id'].astype('int32')).zfill(6)
        folder = fileName[0:-3]
        musicPath = f'{musicSrcPath}/{folder}/{fileName}.mp3'
        cat = idx['genre_id'].astype('int32')
        data['category'].append(cat)
        data['path'].append(musicPath)
        if not idx['genre_id'].astype('int32'):
            continue
        if csvName == 'none':
            csvName = genreMap.get(cat , 'none')
    pd.DataFrame(data=data).to_csv(f'{savePath}/{csvName}.csv', index=False)
    print('done', csvName)
```
